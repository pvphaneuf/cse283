```python
import numpy as np
```

```python
alpha = np.array(0.5)
F_0 = np.array([1, 0, 0, 0, 0, 1])
A = np.array([[0, 1/2, 0, 1/2, 0, 0],
             [1/2, 0, 0, 1/2, 0, 0],
             [0, 0, 0, 0, 1/2, 1/2],
             [1/3, 1/3, 0, 0, 0, 1/3],
             [0, 0, 1/2, 0, 0, 1/2],
             [0, 0, 1/3, 1/3, 1/3, 0]])
F_t = F_0.copy()
```

```python
def get_next_F_t(current_F, initial_F, A, alpha, decimals):
    next_F_t = alpha.dot(current_F).dot(A) + (1 - alpha) * initial_F
    next_F_t = np.around(next_F_t, decimals = decimals)
    return next_F_t
```

```python
print(F_t)
for _ in range(20):
    F_t = get_next_F_t(F_t, F_0, A, alpha, 3)
    print(F_t)
```

```python
print(F_t)
for _ in range(20):
    F_t = get_next_F_t(F_t, F_0, A, alpha, 4)
    print(F_t)
```
