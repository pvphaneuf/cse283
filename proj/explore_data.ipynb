### Check what patients are shared between Hiseq RNA data and clinical data, but not in GA RNA data.

```python
trunc_hiseq_rna_data = pd.read_table('data/trunc_illuminahiseq_rna_data.tsv', header=0, skiprows=[1], index_col=0)
```

```python
trunc_ga_rna_data = pd.read_table('data/trunc_illuminaga_rna_data.tsv', header=0, skiprows=[1], index_col=0)
```

```python
clinical_data = pd.read_table('data/clinical/COADREAD.clin.merged.txt', index_col=0)
```

```python
clinical_patient_list = []
column_header_list = list(clinical_data.columns.values)
for column_header in column_header_list:
    clinical_patient_list.append(clinical_data.loc["patient.bcr_patient_barcode"][column_header].upper()) 
    
clinical_patient_set = set(clinical_patient_list)
```

```python
hiseq_patient_set = set(list(trunc_hiseq_rna_data.columns.values))
```

```python
ga_patient_set = set(list(trunc_ga_rna_data.columns.values))
```

```python
hiseq_clinical_shared_set = hiseq_patient_set & clinical_patient_set
len(hiseq_clinical_shared_set)
```

```python
unique_hiseq_clinical_shared_set = hiseq_clinical_shared_set - ga_patient_set
len(unique_hiseq_clinical_shared_set)
```
