```python
import pandas as pd
```

```python
illuminaga_rna_data = pd.read_table('data/COADREAD.rnaseqv2__illuminaga_rnaseqv2__unc_edu__Level_3__RSEM_genes_normalized__data.data.txt', header=0, skiprows=[1], index_col=0)
```

```python
header_list = list(illuminaga_rna_data.columns.values)
rename_dict = {}
for header in header_list:
    rename_dict[header] = header[:12]
illuminaga_rna_data = illuminaga_rna_data.rename(columns=rename_dict)
```
